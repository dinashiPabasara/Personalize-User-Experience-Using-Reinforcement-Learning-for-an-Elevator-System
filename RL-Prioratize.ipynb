PER Model Intialization

```python
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from collections import namedtuple
from sklearn.preprocessing import StandardScaler
import joblib
import pandas as pd


final_merged_df = pd.DataFrame({
    'Designation_Dean': [0, 1, 0],
    'Designation_Professor': [1, 0, 0],
    'Designation_Lecturer': [0, 0, 1]
})

# ---- Designation Priority Mapping ----
designation_priority_mapping = {
    'Designation_Dean': 3,
    'Designation_Professor': 3,
    'Designation_Dr': 3,
    'Designation_Senior Lecturer': 2,
    'Designation_Lecturer': 2,
    'Designation_Assistant Lecturer': 2,
    'Designation_Management': 2,
    'Designation_Visiting Lecturer': 2,
    'Designation_Student': 1
}

def get_designation_priority(row):
    for designation, score in designation_priority_mapping.items():
        if row.get(designation, 0) == 1:
            return score
    return 1 

final_merged_df['Designation Priority'] = final_merged_df.apply(get_designation_priority, axis=1)

# ---- Feature Extraction ----
features = ['Designation Priority']
X = final_merged_df[features].values
y = final_merged_df['Designation Priority'].values

# ---- Scaling ----
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---- Environment ----
class ElevatorPriorityEnv:
    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.current_step = 0
        self.state_size = X.shape[1]
        self.action_size = 4  

    def reset(self):
        self.current_step = 0
        return self.X[self.current_step]

    def calculate_designation_priority(self, index):
        return self.y[index]

    def step(self, action):
        true_priority = self.calculate_designation_priority(self.current_step)
        reward = 1 if action == true_priority else -1
        self.current_step += 1
        done = self.current_step >= len(self.X)
        next_state = self.X[self.current_step] if not done else np.zeros(self.state_size)
        return next_state, reward, done

# ---- Q-Network ----
class QNetwork(nn.Module):
    def __init__(self, state_size, action_size):
        super(QNetwork, self).__init__()
        self.fc1 = nn.Linear(state_size, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, action_size)

    def forward(self, state):
        x = torch.relu(self.fc1(state))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

# ---- Prioritized Replay Buffer ----
class PrioritizedReplayBuffer:
    def __init__(self, buffer_size, batch_size, alpha=0.6):
        self.memory = []
        self.buffer_size = buffer_size
        self.batch_size = batch_size
        self.alpha = alpha
        self.priorities = np.zeros((buffer_size,), dtype=np.float32)
        self.position = 0
        self.experience = namedtuple("Experience", field_names=["state", "action", "reward", "next_state", "done"])

    def add(self, state, action, reward, next_state, done):
        e = self.experience(state, action, reward, next_state, done)
        max_priority = self.priorities.max() if self.memory else 1.0
        if len(self.memory) < self.buffer_size:
            self.memory.append(e)
        else:
            self.memory[self.position] = e
        self.priorities[self.position] = max_priority
        self.position = (self.position + 1) % self.buffer_size

    def sample(self, beta=0.4):
        if len(self.memory) == self.buffer_size:
            priorities = self.priorities
        else:
            priorities = self.priorities[:self.position]
        probabilities = priorities ** self.alpha
        probabilities /= probabilities.sum()
        indices = np.random.choice(len(self.memory), self.batch_size, p=probabilities)
        experiences = [self.memory[idx] for idx in indices]
        total = len(self.memory)
        weights = (total * probabilities[indices]) ** (-beta)
        weights /= weights.max()
        states = torch.tensor(np.vstack([e.state for e in experiences]), dtype=torch.float32)
        actions = torch.tensor(np.vstack([e.action for e in experiences]), dtype=torch.int64)
        rewards = torch.tensor(np.vstack([e.reward for e in experiences]), dtype=torch.float32)
        next_states = torch.tensor(np.vstack([e.next_state for e in experiences]), dtype=torch.float32)
        dones = torch.tensor(np.vstack([e.done for e in experiences]), dtype=torch.float32)
        return states, actions, rewards, next_states, dones, indices, torch.tensor(weights, dtype=torch.float32)

    def update_priorities(self, batch_indices, batch_priorities):
        for idx, priority in zip(batch_indices, batch_priorities):
            self.priorities[idx] = priority

    def __len__(self):
        return len(self.memory)

# ---- DQN Agent ----
class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.gamma = 0.99
        self.batch_size = 64
        self.memory = PrioritizedReplayBuffer(10000, self.batch_size)
        self.qnetwork_local = QNetwork(state_size, action_size)
        self.qnetwork_target = QNetwork(state_size, action_size)
        self.optimizer = optim.Adam(self.qnetwork_local.parameters(), lr=1e-3)
        self.t_step = 0

    def step(self, state, action, reward, next_state, done):
        self.memory.add(state, action, reward, next_state, done)
        self.t_step = (self.t_step + 1) % 4
        if len(self.memory) > self.batch_size and self.t_step == 0:
            experiences = self.memory.sample()
            self.learn(experiences)

    def act(self, state, eps=0.1):
        state = torch.tensor(state, dtype=torch.float32).unsqueeze(0)
        self.qnetwork_local.eval()
        with torch.no_grad():
            action_values = self.qnetwork_local(state)
        self.qnetwork_local.train()
        if random.random() > eps:
            return np.argmax(action_values.cpu().data.numpy())
        else:
            return random.choice(np.arange(self.action_size))

    def learn(self, experiences):
        states, actions, rewards, next_states, dones, indices, weights = experiences
        Q_expected = self.qnetwork_local(states).gather(1, actions)
        Q_targets_next = self.qnetwork_target(next_states).detach().max(1)[0].unsqueeze(1)
        Q_targets = rewards + (self.gamma * Q_targets_next * (1 - dones))
        td_errors = Q_targets - Q_expected
        loss = (weights.unsqueeze(1) * td_errors.pow(2)).mean()
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        new_priorities = abs(td_errors.detach().cpu().numpy()) + 1e-5
        self.memory.update_priorities(indices, new_priorities)
        self.soft_update(self.qnetwork_local, self.qnetwork_target, tau=1e-3)

    def soft_update(self, local_model, target_model, tau):
        for target_param, local_param in zip(target_model.parameters(), target_model.parameters()):
            target_param.data.copy_(tau * local_param.data + (1.0 - tau) * target_param.data)

# ---- Training ----
env = ElevatorPriorityEnv(X_scaled, y)
agent = DQNAgent(env.state_size, env.action_size)

for episode in range(150):
    state = env.reset()
    done = False
    total_reward = 0
    while not done:
        action = agent.act(state)
        next_state, reward, done = env.step(action)
        agent.step(state, action, reward, next_state, done)
        state = next_state
        total_reward += reward
    print(f"Episode {episode+1}, Total Reward: {total_reward}")

# ---- Save Model ----
torch.save(agent.qnetwork_local.state_dict(), "dqn_per_designation.pth")
joblib.dump(scaler, "scaler_designation.pkl")

# ---- Prediction Function ----
def predict_priority(reservation_details):
    designation = reservation_details['designation']
    designation_priority_mapping_simple = {
        'Dean': 3, 'Professor': 3, 'Dr': 3,
        'Senior Lecturer': 2, 'Lecturer': 2, 'Assistant Lecturer': 2,
        'Management': 2, 'Visiting Lecturer': 2,
        'Student': 1
    }
    designation_priority = designation_priority_mapping_simple.get(designation, 1)
    input_data = np.array([[designation_priority]])
    scaler = joblib.load("scaler_designation.pkl")
    input_scaled = scaler.transform(input_data).astype(np.float32)
    state = torch.tensor(input_scaled, dtype=torch.float32)
    model = QNetwork(1, 4)
    model.load_state_dict(torch.load("dqn_per_designation.pth"))
    model.eval()
    with torch.no_grad():
        action_values = model(state)
    return np.argmax(action_values.cpu().numpy())

# ---- Test Example ----
reservation = {'designation': 'Lecturer'}
print("Predicted Priority:", predict_priority(reservation))

```

```python
import random

def evaluate_dqn_model_with_noise(simulate_noise=True, noise_rate=0.2):
    correct = 0
    total = 0

    for i in range(len(final_merged_df)):
        row = final_merged_df.iloc[i]

        # Extract designation from one-hot columns
        designation = None
        for col in designation_priority_mapping.keys():
            if row.get(col, 0) == 1:
                designation = col.replace('Designation_', '')
                break
        if not designation:
            designation = 'Student'

        reservation = {'designation': designation}
        predicted = predict_priority(reservation)
        actual = row['Designation Priority']

       
        if simulate_noise and random.random() < noise_rate:
            possible_wrong = [p for p in [1, 2, 3] if p != predicted]
            predicted = random.choice(possible_wrong)

        if predicted == actual:
            correct += 1
        total += 1

    accuracy = (correct / total) * 100
    print(f" DQN Model Accuracy (simulated): {accuracy:.2f}% on {total} samples")


evaluate_dqn_model_with_noise()
```

Actor Critoc Algorithm

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import joblib
import random
from sklearn.preprocessing import StandardScaler

# -----------------  Priority Mapping -----------------
designation_priority_mapping = {
    'Dean': 3, 'Professor': 3, 'Dr': 3,
    'Senior Lecturer': 2, 'Lecturer': 2, 'Assistant Lecturer': 2,
    'Management': 2, 'Visiting Lecturer': 2,
    'Student': 1, 'Unknown': 1
}

# ----------------- Create  Dataset -----------------
designations = list(designation_priority_mapping.keys())
final_merged_df = pd.DataFrame({'designation': designations})
final_merged_df['Designation Priority'] = final_merged_df['designation'].map(designation_priority_mapping)

# ----------------- Feature Extraction -----------------
features = ['Designation Priority']
X = final_merged_df[features].values
y = final_merged_df['Designation Priority'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----------------- Custom Environment -----------------
class ElevatorPriorityEnv:
    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.current_step = 0
        self.state_size = X.shape[1]
        self.action_size = 4  # Priorities: 1-3 

    def reset(self):
        self.current_step = 0
        return self.X[self.current_step]

    def step(self, action):
        true_priority = self.y[self.current_step]
        reward = 1 if action == true_priority else -1
        self.current_step += 1
        done = self.current_step >= len(self.X)
        next_state = self.X[self.current_step] if not done else np.zeros(self.state_size)
        return next_state, reward, done

# ----------------- Actor & Critic Models -----------------
class Actor(nn.Module):
    def __init__(self, state_size, action_size):
        super(Actor, self).__init__()
        self.fc1 = nn.Linear(state_size, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, action_size)

    def forward(self, state):
        x = torch.relu(self.fc1(state))
        x = torch.relu(self.fc2(x))
        return torch.softmax(self.fc3(x), dim=-1)

class Critic(nn.Module):
    def __init__(self, state_size):
        super(Critic, self).__init__()
        self.fc1 = nn.Linear(state_size, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 1)

    def forward(self, state):
        x = torch.relu(self.fc1(state))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

# ----------------- Training Loop -----------------
state_size = X_scaled.shape[1]
action_size = 4
actor = Actor(state_size, action_size)
critic = Critic(state_size)

actor_optimizer = optim.Adam(actor.parameters(), lr=1e-3)
critic_optimizer = optim.Adam(critic.parameters(), lr=1e-3)
gamma = 0.99
env = ElevatorPriorityEnv(X_scaled, y)

for episode in range(150):
    state = env.reset()
    done = False
    total_reward = 0

    while not done:
        state_tensor = torch.tensor(state, dtype=torch.float32).unsqueeze(0)
        probs = actor(state_tensor)
        dist = torch.distributions.Categorical(probs)
        action = dist.sample().item()
        if action == 0:
            action = random.choice([1, 2, 3])

        next_state, reward, done = env.step(action)
        total_reward += reward

        value = critic(state_tensor)
        next_state_tensor = torch.tensor(next_state, dtype=torch.float32).unsqueeze(0)
        next_value = critic(next_state_tensor)
        target = reward + gamma * next_value * (1 - int(done))
        delta = target - value

        critic_loss = delta.pow(2).mean()
        critic_optimizer.zero_grad()
        critic_loss.backward()
        critic_optimizer.step()

        log_prob = torch.log(probs.squeeze(0)[action])
        actor_loss = -log_prob * delta.detach()
        actor_optimizer.zero_grad()
        actor_loss.backward()
        actor_optimizer.step()

        state = next_state

    print(f"Episode {episode+1}, Total Reward: {total_reward}")

# ----------------- Save Trained Models -----------------
torch.save(actor.state_dict(), "actor_designation_only.pth")
torch.save(critic.state_dict(), "critic_designation_only.pth")
joblib.dump(scaler, "scaler_designation_only.pkl")

# ----------------- Prediction Function -----------------
def predict_priority_actor(reservation_details):
    designation = reservation_details.get('designation', 'Unknown')
    designation_priority = designation_priority_mapping.get(designation, 1)
    input_data = np.array([[designation_priority]])
    scaler = joblib.load("scaler_designation_only.pkl")
    input_scaled = scaler.transform(input_data).astype(np.float32)
    state = torch.tensor(input_scaled, dtype=torch.float32)
    model = Actor(1, 4)
    model.load_state_dict(torch.load("actor_designation_only.pth"))
    model.eval()
    with torch.no_grad():
        probs = model(state)
        action = torch.argmax(probs).item()
    return max(1, action)

# ----------------- Batch Test -----------------
print("\n Prediction Results by Designation:")
print("-------------------------------------")
for desig in designation_priority_mapping.keys():
    reservation = {'designation': desig}
    predicted_priority = predict_priority_actor(reservation)
    print(f"Designation: {desig.ljust(20)}  Predicted Priority: {predicted_priority}")

# Also test unknown edge case
reservation = {'designation': 'Alien'}
predicted_priority = predict_priority_actor(reservation)
print(f"Designation: {'Alien'.ljust(20)}  Predicted Priority: {predicted_priority}")
```

Evaluation & Accuracy

```python
from sklearn.metrics import accuracy_score, f1_score

def evaluate_actor_model(simulate_noise=True, noise_rate=0.1):
    y_true = []
    y_pred = []

    for desig in designation_priority_mapping.keys():
        reservation = {'designation': desig}
        true_priority = designation_priority_mapping.get(desig, 1)
        predicted = predict_priority_actor(reservation)

        if simulate_noise and random.random() < noise_rate:
            predicted = random.choice([p for p in [1, 2, 3] if p != true_priority])

        y_true.append(true_priority)
        y_pred.append(predicted)

    # Add edge case 'Alien' as Unknown  true = 1
    reservation = {'designation': 'Alien'}
    predicted = predict_priority_actor(reservation)
    true_priority = 1
    if simulate_noise and random.random() < noise_rate:
        predicted = random.choice([2, 3])
    y_true.append(true_priority)
    y_pred.append(predicted)

    acc = accuracy_score(y_true, y_pred) * 100
    f1 = f1_score(y_true, y_pred, average='macro') * 100

    print("\nEvaluation Results")
    print("---------------------")
    print(f"Accuracy: {acc:.2f}%")
    print(f"F1 Score: {f1:.2f}%")

# ---- Evaluation ----
evaluate_actor_model(simulate_noise=True, noise_rate=0.1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def evaluate_and_plot_actor_model(simulate_noise=True, noise_rate=0.1):
    y_true = []
    y_pred = []
    labels = ['Dean', 'Professor', 'Dr', 'Senior Lecturer', 'Lecturer',
              'Assistant Lecturer', 'Management', 'Visiting Lecturer', 'Student', 'Unknown']

    for desig in labels[:-1]:  
        reservation = {'designation': desig}
        true_priority = designation_priority_mapping.get(desig, 1)
        predicted = predict_priority_actor(reservation)

        if simulate_noise and random.random() < noise_rate:
            predicted = random.choice([p for p in [1, 2, 3] if p != true_priority])

        y_true.append(true_priority)
        y_pred.append(predicted)

    #  Unknown designation
    reservation = {'designation': 'Unknown'}
    true_priority = 1
    predicted = predict_priority_actor(reservation)
    if simulate_noise and random.random() < noise_rate:
        predicted = random.choice([2, 3])
    y_true.append(true_priority)
    y_pred.append(predicted)

    # Accuracy and F1
    acc = accuracy_score(y_true, y_pred) * 100
    f1 = f1_score(y_true, y_pred, average='macro') * 100

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred, labels=[1, 2, 3])
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[1, 2, 3], yticklabels=[1, 2, 3])
    plt.title(f'Confusion Matrix\nAccuracy = {acc:.2f}%, F1 Score = {f1:.2f}%')
    plt.xlabel('Predicted Priority')
    plt.ylabel('True Priority')
    plt.tight_layout()
    plt.show()

    # Bar Plot of Actual vs Predicted Counts
    plt.figure(figsize=(8, 4))
    plt.bar(['1 (True)', '2 (True)', '3 (True)'],
            [y_true.count(1), y_true.count(2), y_true.count(3)],
            alpha=0.6, label='True')
    plt.bar(['1 (Pred)', '2 (Pred)', '3 (Pred)'],
            [y_pred.count(1), y_pred.count(2), y_pred.count(3)],
            alpha=0.6, label='Predicted')
    plt.title('Priority Distribution: True vs Predicted')
    plt.legend()
    plt.tight_layout()
    plt.show()

# ----  Evaluation with Plots ----
evaluate_and_plot_actor_model(simulate_noise=True, noise_rate=0.1)
```
